# Análisis del Conjunto de Datos de Precios de Aguacate

**Conjunto de Datos de Precios de Aguacate**: El conjunto de datos "Precios de Aguacate", obtenido de Kaggle, es un conjunto de datos ampliamente utilizado para proyectos de análisis de datos y aprendizaje automático. Proporciona datos históricos sobre precios y ventas de aguacates en varias regiones de los Estados Unidos. Este conjunto de datos es valioso para entender las tendencias en los precios de los aguacates, los volúmenes de ventas y su relación con diferentes factores.

## Atributos Clave

- **Columnas**: El conjunto de datos incluye varias columnas de información. Algunas de las columnas clave típicamente encontradas en este conjunto de datos incluyen:
    - **Fecha** (`Date`): La fecha de observación.
    - **Precio Promedio** (`AveragePrice`): El precio promedio de los aguacates.
    - **Volumen Total** (`Total Volume`): El volumen total de aguacates vendidos.
    - **4046**: Volumen de aguacates Hass pequeños vendidos.
    - **4225**: Volumen de aguacates Hass grandes vendidos.
    - **4770**: Volumen de aguacates Hass extra grandes vendidos.
    - **Bolsas Totales** (`Total Bags`): Total de bolsas de aguacates vendidas.
    - **Bolsas Pequeñas** (`Small Bags`): Bolsas de aguacates pequeños vendidas.
    - **Bolsas Grandes** (`Large Bags`): Bolsas de aguacates grandes vendidas.
    - **Bolsas Extra Grandes** (`XLarge Bags`): Bolsas de aguacates extra grandes vendidas.
    - **Tipo** (`Type`): El tipo de aguacates, generalmente categorizados como convencionales u orgánicos.
    - **Región** (`Region`): La región o ciudad dentro de los Estados Unidos donde se registraron los datos.

- **Rango de Fechas**: El conjunto de datos abarca un rango de fechas, lo que permite el análisis de series de tiempo. Puedes examinar cómo cambian los precios y ventas de aguacates a lo largo de diferentes estaciones y años.

- **Regiones**: Se proporciona información para varias regiones o ciudades a través de los Estados Unidos, lo que permite el análisis de variaciones de precios y ventas en diferentes mercados.

- **Tipos**: El conjunto de datos distingue entre diferentes tipos de aguacates, como convencionales y orgánicos, lo que puede ser útil para comparar tendencias de precios entre estas categorías.

- **Volumen**: Están disponibles datos sobre el volumen total de aguacates vendidos. Esta métrica de volumen se utiliza a menudo para analizar la demanda del mercado.

- **Precio Promedio**: El conjunto de datos contiene el precio promedio de los aguacates, una métrica fundamental para entender las tendencias de precios.

### 3. **Elasticidad del Precio**
**Resumen:** El análisis de elasticidad precio-demanda permite evaluar cómo los cambios en los precios afectan la demanda de aguacates. Comprender la elasticidad puede ayudar a formular estrategias de precios más efectivas.

La fórmula de elasticidad precio-demanda es:

$$
E_d = \frac{\% \text{Cambio en la cantidad demandada}}{\% \text{Cambio en el precio}} = \frac{\Delta Q / Q}{\Delta P / P}
$$

1. **Elasticidad Precio-Demanda por Año:**

   - **Esperado:** Calcula la elasticidad del precio de la demanda para cada año.

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Ruta del archivo
file_path = r"/content/drive/My Drive/aguacates/avocado.csv"  # Usa r"" para tratarlo como raw string

# Cargar el archivo CSV
data = pd.read_csv(file_path)

# Vista previa de los datos
print(data.head())
print(data.info())

```

```python
# Convertir la columna Date a tipo datetime
data['Date'] = pd.to_datetime(data['Date'])

# Comprobar datos nulos
print(data.isnull().sum())
```

```python
# Agrupar por fecha y calcular el promedio de AveragePrice
time_series = data.groupby('Date')['AveragePrice'].mean()

# Verificar la serie temporal
print(time_series.head())
```

```python
# Realizar la descomposición
decompose_result = seasonal_decompose(time_series, model='additive', period=52)

# Visualizar los componentes
decompose_result.plot()

# Establecer que el formato de las fechas muestre primero el mes y luego el año
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m-%Y'))

# Ajustar las etiquetas en el eje X para mayor legibilidad
plt.xticks(rotation=45)  # Rota las etiquetas de los meses y años
plt.show()

```

```python
# Boxplot por tipo
sns.boxplot(x='type', y='AveragePrice', data=data)
plt.title('Precio promedio por tipo de aguacate')
plt.show()
```

```python
# Calcular el precio promedio por región
region_prices = data.groupby('region')['AveragePrice'].mean().sort_values(ascending=False)

# Graficar las regiones con los precios más altos
region_prices.head(10).plot(kind='bar', color='teal')
plt.title('Precio promedio por región (Top 10)')
plt.ylabel('Precio Promedio')
plt.show()
```

```python
# Extraer el año de la columna Date
data['Year'] = data['Date'].dt.year

# Volumen total por año
yearly_volume = data.groupby('Year')['Total Volume'].sum()

# Graficar
yearly_volume.plot(kind='line', marker='o', color='green')
plt.title('Volumen total de aguacates vendidos por año')
plt.xlabel('Año')
plt.ylabel('Volumen Total')
plt.show()
```

```python
# Agrupar por región y fecha, calculando el promedio de AveragePrice
region_seasonality = data.groupby(['region', 'Date'])['AveragePrice'].mean().reset_index()

# Seleccionar algunas regiones para simplificar la visualización
selected_regions = ['California', 'NewYork', 'Texas', 'Florida']
filtered_data = region_seasonality[region_seasonality['region'].isin(selected_regions)]

# Crear un gráfico de líneas para cada región
plt.figure(figsize=(12, 6))
for region in selected_regions:
    regional_data = filtered_data[filtered_data['region'] == region]
    plt.plot(regional_data['Date'], regional_data['AveragePrice'], label=region)

plt.title('Variación de Precios por Región a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Precio Promedio')
plt.legend(title='Región')
plt.grid()
plt.show()
```

```python
# Agrupar por mes y calcular el promedio de AveragePrice
monthly_prices = data.groupby(pd.Grouper(key='Date', freq='M'))['AveragePrice'].mean()

# Graficar precios promedio mensuales
plt.figure(figsize=(12, 6))
plt.plot(monthly_prices.index, monthly_prices.values, color='orange', marker='o')
plt.title('Precios Promedio Mensuales de Aguacates')
plt.xlabel('Mes')
plt.ylabel('Precio Promedio')
plt.grid()
plt.show()
```

```python
# Agrupar por fecha y sumar el volumen total
daily_sales = data.groupby('Date')['Total Volume'].sum()

# Graficar la tendencia del volumen total de ventas
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, daily_sales.values, color='green')
plt.title('Tendencia de Ventas Totales a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Volumen Total de Ventas')
plt.grid()
plt.show()
```

```python
# Agrupar por año y calcular el promedio de AveragePrice
annual_prices = data.groupby(data['Date'].dt.year)['AveragePrice'].mean()

# Graficar los precios promedio anuales
plt.figure(figsize=(10, 5))
plt.bar(annual_prices.index, annual_prices.values, color='purple')
plt.title('Cambios en Precios Anuales Promedio')
plt.xlabel('Año')
plt.ylabel('Precio Promedio')
plt.grid(axis='y')
plt.show()
```

```python
#3.1
# Agrupar por año y calcular el precio promedio y volumen total anual
annual_data = data.groupby('Year').agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'
}).reset_index()

# Calcular las variaciones porcentuales
annual_data['Price Change (%)'] = annual_data['AveragePrice'].pct_change()
annual_data['Volume Change (%)'] = annual_data['Total Volume'].pct_change()

# Calcular la elasticidad precio-demanda
annual_data['Elasticity'] = annual_data['Volume Change (%)'] / annual_data['Price Change (%)']

# Graficar la elasticidad por año
plt.figure(figsize=(10, 6))
plt.plot(annual_data['Year'], annual_data['Elasticity'], marker='o', color='blue')

plt.title('Elasticidad Precio-Demanda por Año', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Elasticidad Precio-Demanda', fontsize=14)
plt.grid(True)
plt.axhline(0, color='red', linestyle='--', linewidth=1)

plt.show()
```

```python
#3.2
# Agrupar por región y año para calcular las elasticidades
region_group = data.groupby(['region', 'Year']).agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'
}).reset_index()

# Calcular las variaciones porcentuales por región
region_group['Price Change (%)'] = region_group.groupby('region')['AveragePrice'].pct_change()
region_group['Volume Change (%)'] = region_group.groupby('region')['Total Volume'].pct_change()

# Calcular elasticidad por región
region_group['Elasticity'] = region_group['Volume Change (%)'] / region_group['Price Change (%)']

# Filtrar la última elasticidad calculada por región
region_elasticity = region_group.groupby('region')['Elasticity'].last().dropna().sort_values()

# Graficar elasticidad por región
plt.figure(figsize=(12, 8))
plt.bar(region_elasticity.index, region_elasticity.values, color='skyblue', edgecolor='black')

plt.title('Elasticidad Precio-Demanda por Región', fontsize=16)
plt.xlabel('Región', fontsize=14)
plt.ylabel('Elasticidad Precio-Demanda', fontsize=14)
plt.xticks(rotation=90)

plt.show()
```

```python
#3.3
# Agrupar por año y calcular volúmenes totales de cada tipo de bolsa
bag_data = data.groupby('Year').agg({
    'AveragePrice': 'mean',
    'Small Bags': 'sum',
    'Large Bags': 'sum',
    'XLarge Bags': 'sum'
}).reset_index()

# Calcular elasticidad para cada tipo de bolsa
for bag_type in ['Small Bags', 'Large Bags', 'XLarge Bags']:
    bag_data[f'{bag_type} Change (%)'] = bag_data[bag_type].pct_change()

bag_data['Elasticity Small'] = bag_data['Small Bags Change (%)'] / bag_data['AveragePrice'].pct_change()
bag_data['Elasticity Large'] = bag_data['Large Bags Change (%)'] / bag_data['AveragePrice'].pct_change()
bag_data['Elasticity XLarge'] = bag_data['XLarge Bags Change (%)'] / bag_data['AveragePrice'].pct_change()

# Graficar elasticidad por tipo de bolsa
elasticities = bag_data[['Elasticity Small', 'Elasticity Large', 'Elasticity XLarge']].mean()
plt.figure(figsize=(8, 6))
plt.bar(elasticities.index, elasticities.values, color=['orange', 'green', 'blue'], edgecolor='black')

plt.title('Elasticidad Precio-Demanda por Tipo de Bolsa', fontsize=16)
plt.xlabel('Tipo de Bolsa', fontsize=14)
plt.ylabel('Elasticidad Precio-Demanda', fontsize=14)

plt.show()
```

```python
#3.4
# Agrupar por tipo (orgánico/convencional) y año
type_group = data.groupby(['type', 'Year']).agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'
}).reset_index()

# Calcular elasticidades para cada tipo
type_group['Price Change (%)'] = type_group.groupby('type')['AveragePrice'].pct_change()
type_group['Volume Change (%)'] = type_group.groupby('type')['Total Volume'].pct_change()
type_group['Elasticity'] = type_group['Volume Change (%)'] / type_group['Price Change (%)']

# Promedio de elasticidades por tipo
type_elasticity = type_group.groupby('type')['Elasticity'].mean()

# Graficar elasticidad por tipo
plt.figure(figsize=(8, 6))
plt.bar(type_elasticity.index, type_elasticity.values, color=['purple', 'green'], edgecolor='black')

plt.title('Elasticidad Precio-Demanda por Tipo de Aguacate', fontsize=16)
plt.xlabel('Tipo de Aguacate', fontsize=14)
plt.ylabel('Elasticidad Precio-Demanda', fontsize=14)

plt.show()
```

```python
#4.5
# Crear cohortes por mes y calcular retención
monthly_cohorts = data.groupby([pd.Grouper(key='Date', freq='M')]).agg({
    'Total Volume': 'sum'
}).reset_index()

# Calcular retención como proporción del volumen del mes inicial
monthly_cohorts['Retention'] = monthly_cohorts['Total Volume'] / monthly_cohorts['Total Volume'].iloc[0]

# Graficar retención
plt.figure(figsize=(10, 6))
plt.plot(monthly_cohorts['Date'], monthly_cohorts['Retention'], marker='o', color='green')

plt.title('Tasa de Retención de Ventas por Cohorte', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Retención (Proporción)', fontsize=14)
plt.grid(True)
plt.axhline(1, color='red', linestyle='--', linewidth=1)

plt.show()
```

```python
#5.3
# Agrupar datos por trimestre y mes
data['Date'] = pd.to_datetime(data['Date'])  # Asegúrate de que la columna Date esté en formato datetime
data['Month'] = data['Date'].dt.month

quarterly_data = data.groupby([pd.Grouper(key='Date', freq='QE-MAR'), 'Month']).agg({
    'AveragePrice': 'mean',
    'Total Volume': 'mean'
}).reset_index()

# Filtrar para los meses del trimestre inicial
quarter1 = quarterly_data[quarterly_data['Month'].isin([1, 2])].copy()

# Simular predicciones usando el promedio de precios del mes siguiente
quarter1['Predicted Price'] = quarter1['AveragePrice'].shift(-1)

# Calcular métricas de evaluación
actual = quarter1['AveragePrice']
predicted = quarter1['Predicted Price']

# Ignorar NaN en las métricas
valid_indices = ~predicted.isna()
rmse = np.sqrt(((actual[valid_indices] - predicted[valid_indices]) ** 2).mean())

print(f'RMSE: {rmse:.4f}')
```

```python
#5.4
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Asegúrate de que la columna 'Date' esté en formato datetime
data['Date'] = pd.to_datetime(data['Date'])

# Filtrar solo columnas numéricas
numeric_columns = data.select_dtypes(include=['number']).columns
data_numeric = data[['Date'] + list(numeric_columns)]

# Agrupar datos por trimestre
quarterly_data = data_numeric.groupby(pd.Grouper(key='Date', freq='QE-MAR')).mean().reset_index()

# Preparar datos para regresión
X = quarterly_data[['Total Volume', '4046', '4225', '4770']].iloc[:-1]
y = quarterly_data['AveragePrice'].iloc[1:]

# Ajustar modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Predicciones
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print(f'R²: {r2}')
print(f'RMSE: {rmse}')
```
